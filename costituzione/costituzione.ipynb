```python
from pprint import pprint
import json
import bs4

file_path = "/content/costituzione.xml"

soup = None

with open(file_path) as file:
  soup = bs4.BeautifulSoup(file.read(), 'html.parser')

capi = soup.find_all('capo')

costituzione = []

for capo in capi:
  capo: bs4.element.Tag = capo
  sezione = capo.find_next('num').text
  articoli = capo.find_all('articolo')

  for articolo in articoli:
    content = ' '.join(map(lambda s: s.text, articolo.find_all('h:p')))
    idArticolo = articolo.get('id')

    costituzione.append({
        'codice': 'costituzione',
        'sezione': sezione,
        'articolo': idArticolo,
        'content': content,
        'metadata_content':
f"""<!--
  metadata:
    codice: costituzione
    sezione: {sezione}
    articolo: {idArticolo}
--!>

{content}"""
    })

pprint(costituzione)

with open('/content/costituzione.json', 'w') as output:
  json.dump(costituzione, output, indent=2)
```
